```python
numbers = pd.Series(np.random.randint(0,1000,10000))
numbers.head()
```

```python
%%timeit -n 1000
total = 0
for number in numbers:
    total+=number
    
total/len(numbers)
```

```python
%%timeit -n 1000
total = np.sum(numbers)
total/len(numbers)

#**Much faster than the iterative method due to vectorization**
```

```python
%%timeit -n 100
s = pd.Series(np.random.randint(0,1000,1000))
for label,value in s.iteritems():
    s.loc[label] = value + 2
```

```python
%%timeit -n 100
s = pd.Series(np.random.randint(0,1000,1000))
s+=2

#**Much faster than the iterative method due to broadcasting**
```

Manipulating DataFrames

```python
df = pd.read_csv('Datasets/presidents.csv')
df.head()
```

```python
def splitname(row):
    row['First'] = row['President'].split(' ')[0]
    row['Last'] = row['President'].split(' ')[-1]
    return row

df=df.apply(splitname,axis='columns')
df.head()
```
